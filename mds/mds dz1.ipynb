Загрузим датасет о ценах на дома в Бостоне. 

```
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import load_boston
import math
from mpl_toolkits.mplot3d import Axes3D

boston = load_boston()
print(boston.data.shape)
print(boston.feature_names)
```

Каждый дом - это вектор в 13-мерном пространстве. 
Написать код для подсчета скалярного произведения двух первых домов из датасета.

```
x1 = boston.data[0] # дом 1
x2 = boston.data[1] # дом 2

```

```
np.dot(x1,x2)
# sum([i*j for i,j in zip(x1, x2)]) 
```

этап 2: взять два атрибута и посчитать скалярное произведение двух векторов (2 дома), начертить 2 дома как вектора в двухмерном пространстве, и сравниваем графически и через скалярное произведение.


```
x = x1[5] , x1[12] #Кладем координаты векторов
y = x2[5] , x2[12]
```

```
fig = plt.figure(figsize=(10, 10))
ax = fig.gca()

ax.plot([0,x[0]], [0,x[1]])
ax.plot([0,y[0]], [0,y[1]])

ax.scatter(x[0], x[1], c='b', marker='o')
ax.scatter(y[0], y[1], c='r', marker='o')

ax.set_xlabel(boston.feature_names[5])
ax.set_ylabel(boston.feature_names[12])


plt.show()

print(np.dot(x,y)/math.sqrt(np.dot(x,x)*np.dot(y,y)))
```

# Задание повышенного уровня.

Задан объект в пространстве 2D. 
Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```
TransformArray = np.array([
    [1, 0, 300],
    [0, 1, 200],
    [0, 0, 1]
])

B = np.transpose(TransformArray.dot(np.transpose(A)))
x = B[:,0]
y = B[:,1]

plt.plot(x, y)
plt.ylim([0, 600])
plt.xlim([0, 600])
```

Уменьшить объект в два раза.

```
TransformArray = np.array([
    [0.5, 0, 0],
    [0, 0.5, 0],
    [0, 0, 1]
])

C = np.transpose(TransformArray.dot(np.transpose(A)))
x = C[:,0]
y = C[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Повернуть на 130 градусов.

```
rad = math.radians(130)

TransformArray = np.array([
    [math.cos(rad), math.sin(rad), 0],
    [-math.sin(rad), math.cos(rad), 0],
    [0, 0, 1]
])

D = np.transpose(TransformArray.dot(np.transpose(A)))
x = D[:,0]
y = D[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1,2)

```
Sme = np.array([
    [1, 0, 200],
    [0, 1, 300],
    [0, 0, 1]
])

Scale = np.array([
    [0.5, 0, 0],
    [0, 1.2, 0],
    [0, 0, 1]
])

B = np.transpose(Scale.dot(Sme.dot(np.transpose(A))))
x = B[:,0]
y = B[:,1]

plt.plot(x, y)
plt.ylim([0, 600])
plt.xlim([0, 600])
```
