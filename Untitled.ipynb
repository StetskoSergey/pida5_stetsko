```python
import pandas as pd

tt = pd.read_csv('obr.csv')
tt['date'] = pd.to_datetime(tt.date, format='%Y-%m-%d')
```

```python
import re

#re.sub("^\s+|\n|\r|\s+$", '', mystr)

def pars_am(row):
    prom = re.sub("^\s+|\n|\xa0|\r|\s+$", '', row.amount)
    prom = re.sub(",", '.', prom)
    return float(prom)
tt['amount'] = tt.apply(pars_am, axis=1)    
```

```python
tt.to_csv('obr.csv')
```

```python
tt = pd.read_csv('obr.csv')
tt['date'] = pd.to_datetime(tt.date, format='%Y-%m-%d')
```

```python
def pars_year(row):
    return row.date.year

tt['year'] = tt.apply(pars_year, axis = 1)
```

```python
def pars_month(row):
    return row.date.month

tt['month'] = tt.apply(pars_month, axis = 1)
```

```python
tt.to_csv('obr.csv')
```
