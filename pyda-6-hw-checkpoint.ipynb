```python
#%cd "C:\Users\Stesk\Documents\GitHub\pida5_stetsko"
import pandas as pd

ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
def class_(row):
    if row['rating'] <= 2:
        return 'низкий рейтинг'
    elif  row['rating'] <= 4:
        return 'средний рейтинг'
    else:
        return 'высокий рейтинг'

    
ratings['class'] = ratings.apply(class_, axis = 1)
ratings.head()
        
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()



```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def region(reyword):
    for region in geo_data:
        for city in geo_data[region]:
            if city in reyword.split():
                return region
    return ''

keywords['region']=keywords['keyword'].apply(region)
keywords[(keywords['region'] != '')].head()
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('ml-latest-small/movies.csv')
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
years = [str(year) for year in range(1950, 2011)]

def production_year(title):
    for year in years:
        if title.find(year)>0:
            return year
    return '1900'

joined['year']=joined['title'].apply(production_year)
joined.head()
```

```python
joined.groupby('year').agg({'rating': 'mean'}).sort_values('rating', ascending=False).head()
```
