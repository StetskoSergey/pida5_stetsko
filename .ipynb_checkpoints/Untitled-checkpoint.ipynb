```python
import pandas as pd

tt = pd.read_csv('obr.csv', sep = ';')
tt['date'] = pd.to_datetime(tt.date, format='%d.%m.%Y')
```

```python
import re

#re.sub("^\s+|\n|\r|\s+$", '', mystr)

def pars_am(row):
    prom = re.sub("^\s+|\n|\xa0|\r|\s+$", '', row.amount)
    prom = re.sub(",", '.', prom)
    return float(prom)
tt['amount'] = tt.apply(pars_am, axis=1)    
```

```python
tt.to_csv('obr.csv')
```
