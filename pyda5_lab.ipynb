```python
def find_slova(bukva, phrasa):
    spisok_slov = phrasa.replace(',','').replace('(','').replace(')','').replace('.','').split()
    return [_ for _ in spisok_slov if bukva == _[-1]]

find_slova('s', string_01)

```

Задача 2
Напишите код, который будет генерировать случайную последовательность из 6
цифр, одна из которых должна быть 3. Позиция цифры 3 должна
определяться случайным образом. Результат должен быть строкой. Например:
"456309" или "330127".

```python
import random

def ch_generate(n, sifra):
    spisok = [str(random.randint(0,9)) for _ in range(n-1)]
    spisok.append(str(sifra))
    random.shuffle(spisok)
    while spisok[0]=='0':
        random.shuffle(spisok)
    return ''.join(spisok)

ch_generate(6, 3)
```
