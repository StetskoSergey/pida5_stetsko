```python
import numpy as np
import pandas as pd
import math
from collections import Counter
```

Функция получения GAIN

```python
def HGini(l):
    length = len(l)
    cnt = Counter(l)
    return sum([i/length*(1-i/length) for i in cnt.values()])
    
def IG(H, l, i):
    return H(l) - i/len(l)*H(l[:i]) - (len(l)-i)/len(l)*H(l[i:])
```

Функция получения максимума из dataset

```python
# получение максимума из столбца
def max_Gain(l):
    key = max(range(len(l)), key = lambda x: IG(HGini, l, x))
    return key , IG(HGini, l, key)

# определение столбца с максимальным GAIN
def max_feature(db):
    feature = max(db.columns, key = lambda x: max_Gain(db[x])[1])
    return feature , max_Gain(db[feature])

# Функция выводящяя дерево решений 
def Result(db):
    feature , i = max_feature(db)
    if i[1] > 0:
        # нужно разделить датасет на две части
        return ( feature, i, Result(db[:i[0]])), ( feature, i, Result(db[i[0]:]))
    else:
        # выводим конечный лист дерева
        return feature , i, db
```

Построим дерево решений на примере датасет диабет

```python
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
X_full, y_full = dataset.data, dataset.target
data = pd.DataFrame(data= np.c_[dataset['data'], dataset['target']],
                     columns= dataset['feature_names'] + ['target'])
```
